```python

daly_active = functions.get_daily_active_users()
```

```python


#
plt.figure(figsize=(12, 6))
plt.plot(daly_active['Date'],daly_active['UniqueActivePlayers'], marker='o')
plt.title("Daily Unique Active Players")
plt.xlabel("Date")
plt.ylabel("Number of Players")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
monthly_active = functions.get_monthly_active_users()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(monthly_active['Month'], monthly_active['UniqueActivePlayers'], marker='o')
plt.title("Monthly Unique Active Players (Line Graph)")
plt.xlabel("Month")
plt.ylabel("Number of Players")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
stickness = functions.get_stickyness()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(stickness['Month'],stickness['Stickiness'], marker='o')
plt.title("Player Stickiness Over Time (DAU / MAU)")
plt.xlabel("Month")
plt.ylabel("Stickiness Ratio")
plt.ylim(0, 1)  
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

- The stickness starts of at 0.2, 20 percent of players where playing monthly 
- As the months go on engaement decliens and even halfs the that 20 percent to 10 percent and even lower in some motns
- players come back monthly but not daily

```python
avg_session_per_user = functions.get_avg_sessions_per_month()
```

```python
level_progress = functions.get_average_progress_speed()
```

```python

plt.figure(figsize=(10, 6))
sns.barplot(x=level_progress.values, y=level_progress.index, palette="viridis")
plt.xlabel('Average Progress per Minute')
plt.ylabel('Level')
plt.title('How Quickly Players Progress Through Levels')
plt.tight_layout()
plt.show()
```

- we can see thta progress is highest in the earlier leves
- with a steady decline 
- this decline becomes less dramatic after the first 8 levels and steadys out as you would expect with a game
- with the last level being having the lowest progression rate
- though the high progression rate could also be attrivuted to player drop off, with most players sticking around for early levels and then not sticking with the game
